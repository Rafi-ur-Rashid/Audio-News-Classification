```python
'''Return train, val and test dataframes according to given proportions'''
# https://stackoverflow.com/questions/24147278/how-do-i-create-test-and-train-samples-from-one-dataframe-with-pandas
def get_train_val_test_dataframes(df_original, proportions=[0.7, 0.1, 0.2], column_filter='channel_names', use_stratify=True):
    # train = df.sample(frac=proportions[0],random_state=RANDOM_STATE) #random state is a seed value
    # test = df.drop(train.index)

    if use_stratify:
        train, remaining = train_test_split(df_original, test_size=1-proportions[0], stratify=df_original[column_filter], random_state=RANDOM_STATE) # here 0.3
    else:
        train, remaining = train_test_split(df_original, test_size=1-proportions[0], random_state=RANDOM_STATE)

    proportion_new = proportions[1] / (proportions[1] + proportions[2])

    if use_stratify:
        test, val = train_test_split(remaining, test_size=proportion_new, stratify=remaining[column_filter], random_state=RANDOM_STATE)
    else:
        test, val = train_test_split(remaining, test_size=proportion_new, random_state=RANDOM_STATE)        

    return train, val, test
```

```python
def print_stats(df_original, use_stratify=True):
    df_train, df_val, df_test = get_train_val_test_dataframes(df_original=df_original, use_stratify=use_stratify)

    print(df_original.shape)

    print(np.unique(df_original['channel_names'], return_counts=False))
    print(np.unique(df_original['channel_names'], return_counts=True))
    print(np.unique(df_original['channel_names'], return_counts=True)[1]/np.sum(np.unique(df_original['channel_names'], return_counts=True)[1]))

    print("----"*10)

    print(df_train.shape)
    print(np.unique(df_train['channel_names'], return_counts=False))
    print(np.unique(df_train['channel_names'], return_counts=True))
    print(np.unique(df_train['channel_names'], return_counts=True)[1]/np.sum(np.unique(df_train['channel_names'], return_counts=True)[1]))

    print("----"*10)

    print(df_val.shape)
    print(np.unique(df_val['channel_names'], return_counts=False))
    print(np.unique(df_val['channel_names'], return_counts=True))
    print(np.unique(df_val['channel_names'], return_counts=True)[1]/np.sum(np.unique(df_val['channel_names'], return_counts=True)[1]))

    print("----"*10)

    print(df_test.shape)
    print(np.unique(df_test['channel_names'], return_counts=False))
    print(np.unique(df_test['channel_names'], return_counts=True))
    print(np.unique(df_test['channel_names'], return_counts=True)[1]/np.sum(np.unique(df_test['channel_names'], return_counts=True)[1]))
```

## Now simply append to dataframes

```python
get_old_file_name = lambda f, l: str(l) + "/" + str(f)
get_new_file_name = lambda ind, l: str(l) + "/" + str(ind) + ".mp3"
columns_required = ['old_file_names', 'labels', 'new_file_names']
```

```python
df_train = pd.DataFrame()
df_train = df_train.append(df_economics_train, ignore_index=True)
df_train = df_train.append(df_international_train, ignore_index=True)
df_train = df_train.append(df_national_train, ignore_index=True)
df_train = df_train.append(df_sports_train, ignore_index=True)
df_train = df_train.append(df_entertainment_train, ignore_index=True)

df_train = df_train.sample(frac=1, random_state=RANDOM_STATE)
df_train = df_train.reset_index()
df_train['new_indices'] = df_train.index + 1
# df_train['new_file_names'] = pd.Series(df_train['new_indices'].values.tolist()).str.join('')
df_train["old_file_names"] = df_train.apply(lambda row : get_old_file_name(row['file_names'], row['labels']), axis=1)
df_train["new_file_names"] = df_train.apply(lambda row : get_new_file_name(row['new_indices'], "train"), axis=1)

df_train = df_train[columns_required]

display(df_train)

file_name = os.path.join(BASE_FOLDER, "train.csv")
df_train.to_csv(file_name, index=False)
```
